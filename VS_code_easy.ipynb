### Env tips

By now we understand that python's strength is also in its libraries. Careful library management will be a entanglement in the early stages but a great advantage as the number and complexity of projects increase.

For this reason various package manager like `anaconda` or `mamba` were born, in order to provide simpler and more inclusive management of the `pip` package.

We have seen that sometimes due to laziness and sometimes due to complexity of dependencies to be solved managing environments with anaconda can be slow.

For this reason now i will propose a set up that uses pip and venv for the reason said before.

We all know that `pip` is the package installer for Python and here we will use it in combination with the module `venv`

The idea is to create a virtual env for each project, save the requirement it in the project folder, and `pip install` it when switching between users. The advantage is that there are no compatibility and portability issues at the expense of the space occupied by the folder with the environment. 

Firstly create a venv:

`python -m venv env_name`

If you have more than one python versione and you want to use a specific one you can:

`py -3.7 -m venv env_name`

At this point the venv module will create a folder called env_name which is the folder to move with the project. It contains the python installation, but currently no other libraries. Generally the lib that you need in order to execute a script are saved in a file called `requirements.txt` that appears like:

In general it is good to specify the version of the libs but it is not necessary.

So now that we have create our `requirements.txt` we can install the desidered libs.

Firstly we activate the env on windows we can:

`env_name\Scripts\activate.bat`

Here you can see that the name of the env appears to the left of the current terminal row. Now we can install the lib like:

`pip install -r requirements.txt`

Here the `-r` is for install from requirements. It wil start the installation and it will take some time. If it will fail is probaly due to the python version.

Once we have created the env and installed the libs we just need to activate it before running our code, but fortunately VS Code comes to our aid by scanning if envs are present and generally allowing us to choose which env to use and save our choise for the future.

VS Code will also detect and let you use conda envs. You can select you env using the command palette or click in the bottom right corner.
